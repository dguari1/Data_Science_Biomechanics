## Lists 

Lists can contain any type of data. Lists are created by using a pair of brackets $[ \ ]$. Elements on the list must be separated by comma

````python
#list that contains the sequence of numbers from 1 to 4  
list_1  = [1, 2, 3, 4] 
print(list_1)
````

````python
#list that contains three strings
list_2 = ["Hello", "Florida", "Tech"]
print(f'My list is : {list_2}')
````

````python
list_2a = ["Hello", "Florida" "Tech"];
````

````python
#list that contains numbers and strings
list_3 = [1, 2, 3, "Hello", "Florida", "Tech" ]
print(list_3)
````

````python
#list that contains lists
list_4 = [list_1, list_2]
print(list_4)
````

## Tuples

Tuples are created by using a pair of parenthesis $( \ )$. Elements on the tuple must be separated by comma

````python
tuple_1 = (1, 2, 3, 2)
tuple_1
````

````python
len(tuple_1)
````

````python
#You can access the elements of a tuple using index
tuple_1[1:4]
````

The diference between **tuples** and **lists** is that 

    tuples -> inmutable 
    list -> mutable 
    
Here is an example:

````python
List = [1,2,3,"Apple","Orange"]
print(List)
List[3] = "Mango"
print(List)
````

````python
Tuple = (1,2,3,"Apple","Orange")
print(Tuple)
Tuple[3] = "Mango"
print(Tuple)
````

````python
list_of_tuples = [('apple', 3), ('banana', 4) , ('orange', 1), ('pear', 4)]
````

````python
list_of_tuples[0]
````

````python
list_of_tuples[1][0]
````

````python
list_of_tuples[0] = ("Mango", 5)
list_of_tuples
````

````python
list_of_tuples[0][0] = "Bean"
````

Tuples and list can be accessed by **unpacking**

````python
a, b, c, d = list_of_tuples
a,b,c,d
````

## Dictionaries

A dictionary is a key-value pairs, and each key maps to a corresponding value. It is defined by using a pair of braces { }, while the elements are a list of comma separated key:value pairs (note the key:value pair is separated by the colon, with key at front and value at the end).

````python
dict_1 = {'apple':3, 'orange':4, 'pear':2}
dict_1
````

To get access to a dictionary, we need to use the key of the element - dictionary[key].

````python
dict_1['apple']
````

````python
dict_1[0] # You cannot access the elements in a dictionary by index!!!!
````

You can access the keys and values of a dictionary using their methods (dictionaries are objects!!)

````python
dict_1.keys()
````

````python
dict_1.values()
````

you can fill a dictionary in multiple ways


1) Creae the dictionary using brackets 
```python
school_dict = {'Florida Tech': 'USA', 'McGill':'Canada'}
school_dict
```

2) Create an empty dictonary fill it by defining the diferent keys and values
```python
school_dict = {}
school_dict['Florida Tec'] = 'USA'
school_dict['McGill'] = 'Canada'
school_dict
```

3) Use the keyword ```dict``` and define a list of tuples
```python
school_dict = dict([('Florida Tech', 'USA'), ('McGill','Canada')])
school_dict
```

you can also query if a key is in the dictionary using the keyword ```in``` or ```not in```

````python
school_dict = {'Florida Tech': 'USA', 'McGill':'Canada'}
'Florida Tech' in school_dict, 'Harvard' in school_dict, 'UC San Diego' not in school_dict
````

You could also use the list function to turn a dictionary with a list of keys. 

````python
keys = list(school_dict)
keys
````
