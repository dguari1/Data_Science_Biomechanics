```python
import sys
from PyQt5 import QtWidgets, QtCore, uic, QtGui
```

```python
class HomeWindow(QtWidgets.QMainWindow):
    def __init__(self, *args, **kwargs):
        super(HomeWindow, self).__init__(*args, **kwargs)

        self.ui = uic.loadUi('test4.ui', self)
        
        #use a validator to make sure that the QLineEdit only accepts int numbers
        self.validator = QtGui.QIntValidator(0,255)
        self.lineEdit_G.setValidator(self.validator)
        self.lineEdit_B.setValidator(self.validator)
        self.lineEdit_R.setValidator(self.validator)
        
        self.pushButton.clicked.connect(self.change_color)
        
    def change_color(self):
        text_B = self.lineEdit_B.text()
        text_G = self.lineEdit_G.text()
        text_R = self.lineEdit_R.text()
        if((text_B) and (text_G) and(text_R)):
            color = 'rgb('+text_R+','+text_G+','+text_B+')'
            self.setStyleSheet('QMainWindow {background-color: '+color+';}')
        
#run the pyqt code        
def main():
    app = QtCore.QCoreApplication.instance()
    if app is None:
        app = QtWidgets.QApplication(sys.argv)

    w = HomeWindow()
    w.setWindowTitle('Test')
    w.show()

    sys.exit(app.exec_())


if __name__ == '__main__':
    main()
```
