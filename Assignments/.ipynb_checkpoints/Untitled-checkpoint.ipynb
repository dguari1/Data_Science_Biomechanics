```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
from scipy.integrate import simps
import numpy as np
import os
```

```python
#list all files in the folder
folder = './DataParkinsons/archive/'
filesToAnalyze = os.listdir(folder)

#these are the columns in the files, including, Time, force for each sensor (8 per foot) and sum of forces (Left and Right)
col_names = ["Time"] + [f"L{i+1}"for i in range(8)] + [f"R{i+1}"for i in range(8)] +['TL', 'TR']


#create an empy dataframe that will store the data for each subject 
subjectData = pd.DataFrame(columns=['duration', 'numberSteps', 'cadence', 'averageStepDuration','variabilityStepDuration', 'correlation', 'state'])


#cover all files that begin by 'Ga', 'Ju' or 'Si'
state = None
for file in filesToAnalyze:
    if ('Ga' in file) or ('Ju' in file) or ('Si' in file):
        #ignore files from dual task experiments, those files include '_10' in the file name
        if '_01.txt' in file:
            
            #idenfify if Patient of Control | Patient -> State = 1 , Control -> state = 0
            if 'Pt' in file:
                state = 1
            else:
                state = 0 
            
            #read the file and process it 
            data = pd.read_csv(folder+file, sep="\t", header=None, names=col_names)
            
            #duration of gait 
            duration = data['Time'].values[-1] - data['Time'].values[0]
            
            #compute the peaks of the force signal, this information will be used to compute several relevant metrics 
            dataSmooth = signal.medfilt(data['TL'],51) #smooth the signal to remove any unwanted peaks
            peaksLeft,_ = signal.find_peaks(dataSmooth,dataSmooth.mean())
            
            dataSmooth = signal.medfilt(data['TR'],51) #smooth the signal to remove any unwanted peaks
            peaksRight,_ = signal.find_peaks(dataSmooth,dataSmooth.mean())
            
            #compute the number of steps
            numberSteps = (len(peaksLeft) + len(peaksRight))/2
            
            #compute the cadence (steps/time) -- average left and right
            cadence = ((len(peaksLeft)/duration) + (len(peaksRight)/duration))/2
            
            #compute the average step duration -- average left and right
            averageStepDuration = ((np.diff(data['Time'][peaksLeft]).mean()) + (np.diff(data['Time'][peaksRight]).mean()))/2
            
            #compute the variability in step duration -- average left and right
            variabilityStepDuration = ((np.diff(data['Time'][peaksLeft]).std()) + (np.diff(data['Time'][peaksRight]).std()))/2
            
            correlation = np.corrcoef(data['TL'], data['TR'])[0,1]
            
            #fill the dataFrame with the data
            subjectData.loc[len(subjectData)] = [duration, numberSteps, cadence, averageStepDuration, variabilityStepDuration,correlation, state]
        else: 
            pass            
    else:
        #ignore the file as it doesn't include force data
        pass

# col_names = ["Time"] + [f"L{i+1}"for i in range(8)] + [f"R{i+1}"for i in range(8)] +['TL', 'TR']
# data = pd.read_csv("./DataParkinsons/archive/GaPt03_01.txt", sep="\t", header=None, names=col_names)
# plt.plot(data['Time'], data['TL'])
# plt.plot(data['Time'], data['TR'])
```

```python
X = subjectData[['duration', 'numberSteps', 'cadence', 'averageStepDuration','variabilityStepDuration', 'correlation']].values
y = subjectData[['state']].values.astype(int).ravel()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

X = subjectData[['duration', 'numberSteps', 'cadence', 'averageStepDuration','variabilityStepDuration', 'correlation']].values

random_seed = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state=random_seed)



rnd_clf = RandomForestClassifier(min_samples_leaf=30, class_weight="balanced")
rnd_clf.fit(X_train, y_train);
ConfusionMatrixDisplay.from_estimator(rnd_clf, X_test,y_test)
plt.show();
```

```python
%matplotlib qt
sf=100
win = 4 * sf
freqs, psd = signal.welch(data['TL'], sf, nperseg=win)


plt.figure(figsize=(8, 4))
plt.plot(freqs, psd, color='k', lw=2)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power spectral density (V^2 / Hz)')
plt.ylim([0, psd.max() * 1.1])
plt.title("Welch's periodogram")
plt.xlim([0, freqs.max()])
```

```python
def get_spectral_feature(signals, sf=100):
    win = 4 * sf
    
    # calcutate the Spectral entropy.
    def spectral_entropy(psd, normalize=False):
        psd_norm = np.divide(psd, psd.sum())
        se = -np.multiply(psd_norm, np.log2(psd_norm)).sum()
        if normalize:
            se /= np.log2(psd_norm.size)
        return se
    
    # calculate the power band for given frequancy.
    def bandpower(psd, freqs, min_freqs, max_freqs):
        # Define delta lower and upper limits
        low, high = min_freqs, max_freqs

        # Find intersecting values in frequency vector
        idx_delta = np.logical_and(freqs >= low, freqs <= high)

        # Frequency resolution
        freq_res = freqs[1] - freqs[0]  # = 1 / 4 = 0.25

        # Compute the absolute power by approximating the area under the curve
        delta_power = simps(psd[idx_delta], dx=freq_res)

        return delta_power/simps(psd, dx=freq_res)

    freqs, psd = signal.welch(signals, sf, nperseg=win)
    #print(dir(psd))
    features = {}
    features["peak_magnitude"] = np.sqrt(psd.max())
    features["entropy"] = spectral_entropy(psd)
    features["half_point"] = freqs.mean()
    
    features["power_bands_0.5_to_1"] = bandpower(psd, freqs, 0.5, 1)
    features["power_bands_0_to_2"] = bandpower(psd, freqs, 0, 2)
    features["power_bands_0_to_4"] = bandpower(psd, freqs, 0, 4)
    features["power_bands_0_to_6"] = bandpower(psd, freqs, 0, 6)
    return features


```
